# Plutopy in action
Here are some practical examples for how to work with plutopy.

```python
import plutopy.data
import matplotlib.pyplot as plt
```

## Get the Pluto image 
The following function retrieves the Pluto mosaic as a GDAL Dataset object. It may take a little while to download the ~300 MB image if this is your first time running plutopy.

```python
pluto = plutopy.data.get_pluto_image()
```

## Plot the Pluto image
The image can be plotted using the `imshow()` function from matplotlib.

```python
image = pluto.GetRasterBand(1).ReadAsArray()
plt.imshow(image)
```
